## From the Golden Ratio to the Golden Angle
The beautiful spiral phyllotaxy of sunflowers and the lateral organs of other plants arises from the Golden Angle. The Golden Angle is derived from the Golden Ratio. If two arcs on a circle are defined such that the length of arc `a` divided by arc `b` is the Golden Ratio, then the resulting angle is the Golden Angle. As we just saw, the Golden Ratio is a product of the Fibonacci sequence, and this is how the inspiring patterns seen in a sunflower disc are linked to the Fibonacci sequence! ðŸŒ»

To approximate the Golden Angle, let's figure out a way to calculate it from the Golden Ratio, as we already know how to do that!

But first, let's get our most accurate estimation of the Golden Ratio.

In the next lesson, we will be using a function called `.pop()`. `.pop()` added to the end of a list will return the last element of the list. As the last member of our list phi is the best approximation of the Golden Ratio that we have, let's use `.pop()` to get that value! The way you use `.pop()` is as follows:

```
my_variable = my_list.pop()
```

In the cell below, create a new variable called `golden_ratio` by using `.pop()` to isolate the last value of the list `phi`. Print out the value of `golden_ratio`.

````python
golden_ratio = phi.pop()
print(golden_ratio)
````

The Golden Ratio has some very interesting properties.

In the cell below, `print()` the following using your new variable `golden_ratio`:

* Print the value of `golden_ratio`
* Print the value of `golden_ratio` squared
* Print the value of `1 / golden_ratio`
  
What do you observe that is similar between all the values that you just calculated?

````python
print(golden_ratio)
print(golden_ratio**2)
print(1/golden_ratio)
````

Not only do all the values you just calculated end in the same fractional component, but also notice that `golden_ratio**2 = 1 + golden_ratio`. With this information, let's figure out how to derive the Golden Angle from the Golden Ratio!

Let `f` equal the fraction of the length of a circle's circumference that is occupied by the arc `b`, where the ratio of the lengths of the arcs `a / b` is the Golden Ratio.

```
golden_ratio = a/b
f = b / (a + b) 
f = (b/b) / ( (a/b) + (b/b) )
f = 1 / (golden_ratio + 1)
f = 1 / golden_ratio**2
```

From the derivation above, you can see that the fraction of the circumference of the circle occupied by the arc `b` that is subtended by the Golden Angle is `1 / golden_ratio**2`!

In the cell below, approximate the Golden Angle from your approximation of the Golden Ratio in degrees as:

````python
golden_angle = 360*(1/golden_ratio**2)
print(golden_angle)
````

In this activity, you will learn how to code the intricate, beautiful packing of florets in a sunflower head (or cactus).

## The formula for a sunflower

```
theta = n * phi
r = sqrt(n)
```

Where `n` is the ordered number of each floret on a sunflower head, `theta` is the angle of the floret, `phi` is the golden angle, and `r` is radius, the floret's distance from the center of the sunflower. You should think of each sequential floret as having one more "turn" of `phi` and having a longer radius than the previous, following a square root function.

Also remember back to trigonometry (nobody's favorite class, but let's try to recall) that in a right triangle, `cos` and `sin` are the adjacent and opposite sides divided by the hypotenuse, respectively. In the context of a radial Cartesian plane centered at $(0,0)$, to retrieve the `x` and `y` coordinates we use the following equations:

```
x = r * cos(theta)
y = r * sin(theta)
```

Finally, we will be working in radians, not degrees. Remember also from trigonometry that to convert degrees into radians, we multiply degrees by $\displaystyle\frac{\pi}{180}$.

# Phyllotaxy as a graph

````python
import matplotlib.pyplot as plt
import math

golden_angle = 137.5077640500378546463487396283702776206886952699253696312384958261062333851951
phi = golden_angle * ( math.pi / 180.0 ) 

def phyllotaxy_graph(number=100, fibo1=5, fibo2=8):

    ###############################
    # calculate phyllotaxy vertices
    ###############################
    
    xlist = []
    ylist = []
    
    for n in range (0, number): 
        r = math.sqrt(n) 
        theta = n * phi 
    
        x = r * math.cos(theta)
        y = r * math.sin(theta)
    
        xlist.append(x)
        ylist.append(y)
    
    ############################
    # calculate phyllotaxy edges
    ############################
    
    # select two consecutive fibonacci numbers as a list
    fibonacci_numbers = [fibo1,fibo2]
    
    # create lists to store lists of parastichy coordinates
    parastichy_xlists = []
    parastichy_ylists = []
    
    # parastichy edges
    for fibo_num in fibonacci_numbers:
        for i in range(fibo_num):
            parastichy_xlists.append(xlist[i::fibo_num])
            parastichy_ylists.append(ylist[i::fibo_num])
    
    ######
    # plot
    ######
    
    for i in range(len(parastichy_xlists)):
        plt.plot(parastichy_xlists[i], parastichy_ylists[i])
    
    plt.gca().set_aspect("equal")
    plt.axis("off")

````

# The Euler Characteristic Transform and phyllotaxy

<img src="images/phyllotaxy_animation.gif" width="800">

````python
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
from ect import ECT, EmbeddedGraph

# Set Euler Characteristic Transform parameters
BOUND_RADIUS = 1
NUM_ECT_DIRECTIONS = 360
ECT_THRESHOLDS = np.linspace(0, BOUND_RADIUS, NUM_ECT_DIRECTIONS)

def create_phyllotaxy_graph(num_vertices, fib_pair):
    """
    Create a phyllotaxy graph compatible with your EmbeddedGraph API.
    
    Returns:
        vertices (Nx2 array)
        G (EmbeddedGraph)
    """
    
    golden_angle = 137.50776405003785
    phi = golden_angle * (math.pi / 180.0)
    
    # --- Generate vertices ---
    vertices = []
    for n in range(num_vertices):
        r = math.sqrt(n)
        theta = n * phi
        x = r * math.cos(theta)
        y = r * math.sin(theta)
        vertices.append([x, y])
    
    vertices = np.array(vertices)
    
    # --- Build EmbeddedGraph correctly ---
    G = EmbeddedGraph()
    
    # Add nodes WITH coordinates (required by your API)
    for i, coord in enumerate(vertices):
        G.add_node(i, coord)
    
    # Add Fibonacci parastichy edges
    for fib in fib_pair:
        for i in range(fib):
            indices = list(range(i, num_vertices, fib))
            for j in range(len(indices) - 1):
                G.add_edge(indices[j], indices[j+1])
    
    return vertices, G

def compute_aligned_polar_ect(G_original):
    """
    Compute polar ECT and aligned embedded graph.
    Clean visualization: no axes, no titles, black graph styling.
    
    Returns:
        fig_graph, fig_ect
    """
    
    # -------------------------
    # 1) FIXED -90Â° ROTATION
    # -------------------------
    angle_rad = np.deg2rad(-90)
    rot_matrix = np.array([
        [np.cos(angle_rad), -np.sin(angle_rad)],
        [np.sin(angle_rad),  np.cos(angle_rad)]
    ])
    
    rotated_coords = G_original.coord_matrix @ rot_matrix.T
    
    G = EmbeddedGraph()
    for node, coord in zip(G_original.nodes(), rotated_coords):
        G.add_node(node, coord)
    
    G.add_edges_from(G_original.edges())
    
    original_coords = G.coord_matrix.copy()
    
    # -------------------------
    # 2) ECT NORMALIZATION
    # -------------------------
    G.center_coordinates(center_type="origin")
    G.transform_coordinates()
    G.scale_coordinates(BOUND_RADIUS)
    
    normalized_coords = G.coord_matrix.copy()
    
    # -------------------------
    # 3) AFFINE RECOVERY
    # -------------------------
    src = np.float32(original_coords[:3])
    dst = np.float32(normalized_coords[:3])
    M = cv2.getAffineTransform(src, dst)
    A = np.vstack([M, [0,0,1]])
    
    homog = np.hstack([rotated_coords, np.ones((len(rotated_coords),1))])
    transformed_vertices = (homog @ A.T)[:,:2]
    
    # -------------------------
    # 4) ECT CALCULATION
    # -------------------------
    ect_calc = ECT(
        num_dirs=NUM_ECT_DIRECTIONS,
        thresholds=ECT_THRESHOLDS,
        bound_radius=BOUND_RADIUS
    )
    
    ect_result = ect_calc.calculate(G)
    
    # -------------------------
    # 5) GRAPH IMAGE (CLEAN)
    # -------------------------
    fig_graph, ax = plt.subplots()
    
    # Black points
    ax.scatter(
        transformed_vertices[:,0],
        transformed_vertices[:,1],
        s=10,
        color='black'
    )
    
    # Black edges
    for i, j in G_original.edges():
        ax.plot(
            [transformed_vertices[i,0], transformed_vertices[j,0]],
            [transformed_vertices[i,1], transformed_vertices[j,1]],
            linewidth=0.5,
            color='black'
        )
    
    ax.set_aspect('equal')
    ax.axis('off')  # removes axes + labels
    plt.close(fig_graph)
    
    # -------------------------
    # 6) POLAR ECT IMAGE (CLEAN)
    # -------------------------
    fig_ect, ax = plt.subplots(subplot_kw=dict(projection="polar"))
    
    thetas = ect_result.directions.thetas
    thresholds = ect_result.thresholds
    THETA, R = np.meshgrid(thetas, thresholds)
    
    ax.pcolormesh(THETA, R, ect_result.T, cmap="inferno")
    
    # Remove ALL polar decorations
    ax.set_axis_off()
    plt.close(fig_ect)
    
    return fig_graph, fig_ect
````

````python
vertices, G = create_phyllotaxy_graph(
    num_vertices=200,
    fib_pair=(21,34)
)

fig_graph, fig_ect = compute_aligned_polar_ect(G)
````

````python
import imageio
import numpy as np
import math
import matplotlib.pyplot as plt
import io
from ect import EmbeddedGraph

def fig_to_array_fixed(fig):
    """
    Convert matplotlib figure to RGB numpy array.
    Uses 'tight' bounding box to eliminate padding around the plot.
    """
    buf = io.BytesIO()
    # pad_inches=0 is key to removing the white border
    fig.savefig(buf, format='png', bbox_inches='tight', pad_inches=0)
    buf.seek(0)
    img = plt.imread(buf)
    
    # Ensure 3-channel RGB
    if img.shape[2] == 4:
        img = img[:, :, :3]
        
    # Convert from float [0,1] if necessary (depends on backend)
    if img.dtype == np.float32 or img.dtype == np.float64:
        img = (img * 255).astype(np.uint8)
        
    return img

def generate_growing_phyllotaxy_gif(
    max_vertices=120,
    step=10,
    output_path="growing_phyllotaxy.gif",
    fps=4
):
    fib_pairs = [(2,3), (5,8), (13,21), (34,55)]
    golden_angle = 137.50776405003785
    phi = golden_angle * (math.pi / 180.0)
    
    frames = []
    
    # Generate the forward sequence
    for N in range(step, max_vertices + 1, step):
        # -----------------------------
        # Build reversed-radius vertices
        # -----------------------------
        thetas = [n * phi for n in range(N)]
        radii = [math.sqrt(n) for n in range(N)]
        radii.reverse()
        
        vertices = []
        for n in range(N):
            r = radii[n]
            theta = thetas[n]
            x = r * math.cos(theta)
            y = r * math.sin(theta)
            vertices.append([x,y])
        
        vertices = np.array(vertices)
        
        # -----------------------------
        # Create master frame figure
        # -----------------------------
        fig = plt.figure(figsize=(16, 8))
        
        for col, fib_pair in enumerate(fib_pairs):
            G = EmbeddedGraph()
            for i, coord in enumerate(vertices):
                G.add_node(i, coord)
            
            for fib in fib_pair:
                for i in range(fib):
                    indices = list(range(i, N, fib))
                    for j in range(len(indices)-1):
                        G.add_edge(indices[j], indices[j+1])
            
            # Compute the two figures via your external function
            fig_graph, fig_ect = compute_aligned_polar_ect(G)
            
            # Convert to arrays WITHOUT padding
            img_graph = fig_to_array_fixed(fig_graph)
            img_ect = fig_to_array_fixed(fig_ect)
            
            plt.close(fig_graph)
            plt.close(fig_ect)
            
            # Top row: graphs
            ax1 = fig.add_subplot(2, 4, col + 1)
            ax1.imshow(img_graph)
            ax1.axis('off')
            
            # Bottom row: ECT
            ax2 = fig.add_subplot(2, 4, col + 5)
            ax2.imshow(img_ect)
            ax2.axis('off')
        
        # -----------------------------
        # Master Layout Adjustments
        # -----------------------------
        plt.subplots_adjust(left=0, right=1, top=1, bottom=0, wspace=0.02, hspace=0.02)
        
        # Capture frame
        fig.canvas.draw()
        frame = np.asarray(fig.canvas.buffer_rgba())[:, :, :3]
        frames.append(frame.copy())
        
        plt.close(fig)
        print(f"Frame N={N} complete")

    # -----------------------------
    # Create Looping & Reverse Logic
    # -----------------------------
    # frames[::-1] reverses the list
    # [1:-1] removes the first and last frames of the reversed list 
    # so we don't have "pauses" caused by duplicate frames at the loop points.
    reverse_sequence = frames[::-1][1:-1]
    full_loop = frames + reverse_sequence
    
    # Save GIF with infinite loop (loop=0)
    # Using imageio v2+ style parameters
    imageio.mimsave(output_path, full_loop, fps=fps, loop=0)
    print(f"\nGIF saved to: {output_path}")
    print(f"Total frames: {len(full_loop)} ({len(frames)} forward, {len(reverse_sequence)} reverse)")

# Run the generation
generate_growing_phyllotaxy_gif(
    max_vertices=100,
    step=6,
    output_path="growing_phyllotaxy.gif",
    fps=5
)
````
